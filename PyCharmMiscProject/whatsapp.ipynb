```python
import pandas as pd
import matplotlib.pyplot as plt
from nltk.sentiment.vader import SentimentIntensityAnalyzer

def sentiment_analysis(data, columnname):
    sid = SentimentIntensityAnalyzer()
    
    # Ensure a copy of the DataFrame to avoid SettingWithCopyWarning
    data = data.copy()
    
    # Drop NaN values
    data.dropna(inplace=True)
    
    # Apply sentiment analysis
    data['scores'] = data[columnname].astype(str).apply(lambda text: sid.polarity_scores(text))
    data['compound'] = data['scores'].apply(lambda score_dict: score_dict['compound'])
    data['negative'] = data['scores'].apply(lambda score_dict: score_dict['neg'])
    data['neutral'] = data['scores'].apply(lambda score_dict: score_dict['neu'])
    data['positive'] = data['scores'].apply(lambda score_dict: score_dict['pos'])
    
    # Classify as 'pos' or 'neg' based on compound score
    data['comp_score'] = data['compound'].apply(lambda c: 'pos' if c >= 0 else 'neg')

    # Count positive and negative scores
    posneg = data['comp_score'].value_counts().to_frame()
    
    # Return the results
    return posneg, data  # Returning both the sentiment count and dataset

# Example usage:
pos, data_senti = sentiment_analysis(dataset, columnname='Chat')

# Display sentiment analysis result
print(pos)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer(max_df=0.95,min_df=2,stop_words='english')
dtm=tfidf.fit_transform(df["Chat"])
```

```python
from sklearn.decomposition import NMF
nmf_model=NMF(n_components=5,random_state=42)
nmf_model.fit(dtm)
```

```python
for index, topic in enumerate(nmf_model.components_):
    results=([tfidf.get_feature_names_out()[i] for i in topic.argsort()[-10:]])
    print(results)
```

```python
topic_results=nmf_model.transform(dtm)
df["Topic"]=topic_results.argmax(axis=1)
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS

# Combine all messages into a single string
text = ' '.join(df['Chat'])

# Define stopwords to remove common words like "the", "is", "and"
stopwords = set(STOPWORDS)
stopwords.update(["media", "omitted", "image", "video", "deleted", "photo"])  # Add more as needed

# Generate the Word Cloud
wordcloud = WordCloud(width=800, height=400, background_color='white', stopwords=stopwords).generate(text)

# Plot the Word Cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")  # Hide axes
plt.title("WhatsApp Chat Word Cloud", fontsize=14)
plt.show()
```
